```python
ww = pd.read_csv("winequality-white.csv",sep=";")
print(ww.describe())
```

```python
# we are creating a new column called "quality_label", we define a range and associate that range with a label
rw['quality_label'] = rw['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')

## here we are transforming these labels into categrical data type (specific to pandas) instead of simple string
rw['quality_label'] = pd.Categorical(rw['quality_label'],
categories=['low', 'medium', 'high'])
```

```python
ww['quality_label'] = ww['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')

my_order = ['low', 'medium', 'high']

ww['quality_label'] = pd.Categorical(ww['quality_label'],
categories=my_order, ordered = True)

#print(ww.head())
```

```python
rw["color"] = "red"
ww["color"] = "white"
#print(ww.head())
comb = pd.concat([rw,ww], ignore_index=True)
print(comb.head())
```

```python
print(comb.tail())
```

```python
pd.concat([rw.describe(), ww.describe()], axis=1, keys=["Red Wines Stats", "White Wines Stats"])
```

```python
corr_matrix_rw = rw.corr()
print(corr_matrix_rw)
corr_matrix_ww = ww.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sn

fig, ax = plt.subplots(1, 2, figsize=(25,10))
ax[0].set_title("Red Wine")
ax[1].set_title("White Wine")
mask1 = np.triu(np.ones_like(corr_matrix_rw, dtype=bool))
mask2 = np.triu(np.ones_like(corr_matrix_ww, dtype=bool))
sn.heatmap(corr_matrix_rw, mask=mask1, annot=True, linewidths=.5, ax=ax[0])
sn.heatmap(corr_matrix_ww, mask=mask2, annot=True, linewidths=.5, ax=ax[1])

plt.show()
```

```python
x= comb["color"].value_counts()["red"]
y= comb["color"].value_counts()["white"]
a=[x,y]
print(a)
z=["red","white"]
colors=sn.color_palette("bright")[0:5]
plt.pie(a,labels=z,colors=colors,autopct="%.0f%%")
plt.show()
```
