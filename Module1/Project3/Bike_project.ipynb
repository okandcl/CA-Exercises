```python
df["dteday"] = pd.to_datetime(df["dteday"])
print(df.dtypes)
```

```python
df.loc[df['holiday'] == 0, 'holiday'] = 'No'
df.loc[df['holiday'] == 1, 'holiday'] = 'Yes'
```

```python
df.loc[df['season'] == 1, 'season'] = 'Winter'
df.loc[df['season'] == 2, 'season'] = 'Spring'
df.loc[df['season'] == 3, 'season'] = 'Summer'
df.loc[df['season'] == 4, 'season'] = 'Fall'
```

```python
df.loc[df['weathersit'] == 1, 'weather'] = 'Clear'
df.loc[df['weathersit'] == 2, 'weather'] = 'Cloudy'
df.loc[df['weathersit'] == 3, 'weather'] = 'Light_Rain'
df.loc[df['weathersit'] == 4, 'weather'] = 'Heavy_Rain'
```

```python
df.loc[df['weekday'] == 0, 'weekday'] = 'Sunday'
df.loc[df['weekday'] == 1, 'weekday'] = 'Monday'
df.loc[df['weekday'] == 2, 'weekday'] = 'Tuesday'
df.loc[df['weekday'] == 3, 'weekday'] = 'Wednesday'
df.loc[df['weekday'] == 4, 'weekday'] = 'Thursday'
df.loc[df['weekday'] == 5, 'weekday'] = 'Friday'
df.loc[df['weekday'] == 6, 'weekday'] = 'Saturday'
```

```python
df.loc[df['yr'] == 0, 'yr'] = '2011'
df.loc[df['yr'] == 1, 'yr'] = '2012'
```

```python
df["temp"] = (df["temp"]*47)-8
```

```python
df["atemp"] = (df["atemp"]*66)-16
```

```python
df["hum"] = df["hum"]*100
```

```python
df["windspeed"] = df["windspeed"]*67
```

```python
display(df.head())
```

```python
df.describe()
```

```python
df["yr"] = df["yr"].astype('category')
df["mnth"] = df["mnth"].astype('category')
df["holiday"] = df["holiday"].astype('category')
df["weekday"] = df["weekday"].astype('category')
df["workingday"] = df["workingday"].astype('category')
df["weathersit"] = df["weathersit"].astype('category')
df["season"] = df["season"].astype('category')
df.info()
```

```python
for i in df.index:
    if df["temp"][i] <= 15:
        df["temp"][i] = "Cold"
    elif df["temp"][i]>15 and df["temp"][i] <= 20:
        df["temp"][i] = "Cool"
    elif df["temp"][i]>20 and df["temp"][i] <= 25:
        df["temp"][i] = "Warm"
    elif df["temp"][i]>25 and df["temp"][i] <= 30:
        df["temp"][i] = "Hot"
    elif df["temp"][i]>30 and df["temp"][i] <= 35:
        df["temp"][i] = "Very_Hot"
    else:
        df["temp"][i] ="Xtrm_Hot"
```

```python
display(df.tail(3000))
```

```python
df["temp"] = df["temp"].astype('category')
df.info()
```

```python
for i in df.index:
    if df["windspeed"][i] <= 5:
        df["windspeed"][i] = "Calm"
    elif df["windspeed"][i]>5 and df["windspeed"][i] <= 25:
        df["windspeed"][i] = "Light"
    elif df["windspeed"][i]>25 and df["windspeed"][i] <= 40:
        df["windspeed"][i] = "Breezy"
    elif df["windspeed"][i]>40 and df["windspeed"][i] <= 50:
        df["windspeed"][i] = "Windy"
    else:
        df["windspeed"][i] ="Very Windy"
```

```python
display(df.tail(3000))
```

```python
df["windspeed"] = df["windspeed"].astype('category')
df.info()
```

```python
sn.set_style('whitegrid')
sn.set_context('talk')
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (30, 10),
          'axes.labelsize': 'x-large',
          'axes.titlesize':'x-large',
          'xtick.labelsize':'x-large',
          'ytick.labelsize':'x-large'}

plt.rcParams.update(params)
```

```python
fig,ax = plt.subplots()
sn.pointplot(data=df[['hr', 'cnt', 'season']],
              x='hr',
              y='cnt',
              hue='season',
              ax=ax)
ax.set(title="Season wise hourly distribution of counts")
```

```python
fig,ax = plt.subplots()
ho = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
sn.pointplot(data=df[['hr',
                           'cnt',
                           'weekday']],
              x='hr',
              y='cnt',
              hue='weekday',
              ax=ax,
              hue_order=ho)
ax.set(title="Weekday wise hourly distribution of counts")
```

```python
fig,axs = plt.subplots(1, 2, sharey="row")
ho = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
sn.lineplot(data=df[['casual', 'hr', 'weekday']],
              x= 'hr',
              y='casual',
              hue='weekday',
              ax=axs[0], hue_order=ho)
sn.lineplot(data=df[['registered', 'hr', 'weekday']],
              x= 'hr',
              y='registered',
              hue='weekday', legend=False,
              ax=axs[1], hue_order=ho)
axs[1].set_ylabel("registered")
plt.show()
```

```python
fig,axs = plt.subplots(1, 2, sharey="row")
sn.lineplot(data=df[['casual', 'hr', 'season']],
              x= 'hr',
              y='casual',
              hue='season',
              ax=axs[0])
sn.lineplot(data=df[['registered', 'hr', 'season']],
              x= 'hr',
              y='registered',
              hue='season', legend=False,
              ax=axs[1])
axs[1].set_ylabel("registered")
plt.show()
```

```python
fig,ax = plt.subplots()
ho = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
sn.pointplot(data=df[['hr',
                           'registered',
                           'weekday']],
              x='hr',
              y='registered',
              hue='weekday',
              ax=ax,
              hue_order=ho)
ax.set(title="regs")
```

```python
fig,ax = plt.subplots()
sn.pointplot(data=df[['hr',
                           'cnt',
                           'holiday']],
              x='hr',
              y='cnt',
              hue='holiday',
              ax=ax)
ax.set(title="regs")
```

```python
fig,axs = plt.subplots(1, 2, sharey="row")
sn.lineplot(data=df[['casual', 'hr', 'holiday']],
              x= 'hr',
              y='casual',
              hue='holiday',
              ax=axs[0])
sn.lineplot(data=df[['registered', 'hr', 'holiday']],
              x= 'hr',
              y='registered',
              hue='holiday', legend=False,
              ax=axs[1])
axs[1].set_ylabel("registered")
plt.show()
```
