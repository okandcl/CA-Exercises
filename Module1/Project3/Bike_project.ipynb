```python
df["dteday"] = pd.to_datetime(df["dteday"])
print(df.dtypes)
```

```python
df.loc[df['holiday'] == 0, 'holiday'] = 'No'
df.loc[df['holiday'] == 1, 'holiday'] = 'Yes'
```

```python
df.loc[df['season'] == 1, 'season'] = 'Winter'
df.loc[df['season'] == 2, 'season'] = 'Spring'
df.loc[df['season'] == 3, 'season'] = 'Summer'
df.loc[df['season'] == 4, 'season'] = 'Fall'
```

```python
df.loc[df['weathersit'] == 1, 'weather'] = 'Clear'
df.loc[df['weathersit'] == 2, 'weather'] = 'Cloudy'
df.loc[df['weathersit'] == 3, 'weather'] = 'Light_Rain'
df.loc[df['weathersit'] == 4, 'weather'] = 'Heavy_Rain'
```

```python
df.loc[df['weekday'] == 0, 'weekday'] = 'Monday'
df.loc[df['weekday'] == 1, 'weekday'] = 'Tuesday'
df.loc[df['weekday'] == 2, 'weekday'] = 'Wednesday'
df.loc[df['weekday'] == 3, 'weekday'] = 'Thursday'
df.loc[df['weekday'] == 4, 'weekday'] = 'Friday'
df.loc[df['weekday'] == 5, 'weekday'] = 'Saturday'
df.loc[df['weekday'] == 6, 'weekday'] = 'Sunday'
```

```python
df.loc[df['yr'] == 0, 'yr'] = '2011'
df.loc[df['yr'] == 1, 'yr'] = '2012'
```

```python
df["temp"] = df["temp"]*41
```

```python
df["atemp"] = df["atemp"]*50
```

```python
df["hum"] = df["hum"]*100
```

```python
df["windspeed"] = df["windspeed"]*67
```

```python
df["yr"] = df["yr"].astype('datetime64[ns]')
df["mnth"] = df["mnth"].astype('category')
df["holiday"] = df["holiday"].astype('category')
df["weekday"] = df["weekday"].astype('category')
df["workingday"] = df["workingday"].astype('category')
df["weathersit"] = df["weathersit"].astype('category')
df["season"] = df["season"].astype('category')
df.info()
```

```python
for i in df.index:
    if df["temp"][i] <= 15:
        df["temp"][i] = "Cold"
    elif df["temp"][i]>15 and df["temp"][i] <= 20:
        df["temp"][i] = "Cool"
    elif df["temp"][i]>20 and df["temp"][i] <= 25:
        df["temp"][i] = "Warm"
    elif df["temp"][i]>25 and df["temp"][i] <= 30:
        df["temp"][i] = "Hot"
    elif df["temp"][i]>30 and df["temp"][i] <= 35:
        df["temp"][i] = "Very_Hot"
    else:
        df["temp"][i] ="Xtrm_Hot"
```
