```python
db_conn = sqlite3.connect("P5.db")
```

```python
c = db_conn.cursor()
```

```python
c.execute(
"""
    CREATE TABLE Customers (
    CustomerID varchar PRIMARY KEY NOT NULL,
    CustomerName varchar NOT NULL,
    Segment varchar NOT NULL
);""")

c.execute("""
CREATE TABLE Adress (
    AdressID varchar PRIMARY KEY NOT NULL,
    City varchar NOT NULL,
    State varchar NOT NULL,
    Region varchar NOT NULL,
    PostalCode integer NOT NULL
);"""
)

c.execute("""
CREATE TABLE OrderDetails (
    OrderID varchar NOT NULL,
    ODKey integer PRIMARY KEY NOT NULL,
    ProductID varchar NOT NULL,
    Quantity integer NOT NULL,
    Sales integer NOT NULL,
    Discount integer NOT NULL,
    Profit integer NOT NULL,
    FOREIGN KEY (OrderID) REFERENCES Orders(OrderID),
    FOREIGN KEY (ProductID) REFERENCES Products(ProductID)
);"""
)

c.execute("""
CREATE TABLE Products (
    ProductID varchar PRIMARY KEY NOT NULL,
    CategoryID varchar NOT NULL,
    SupplierPrice integer NOT NULL,
    CataloguePrice integer NOT NULL,
    ProductName varchar(200) NOT NULL,
    FOREIGN KEY (CategoryID) REFERENCES Category(CategoryID)
);"""
)

c.execute("""
CREATE TABLE Orders (
    OrderID varchar PRIMARY KEY NOT NULL,
    CustomerID varchar NOT NULL,
    OrderDate date NOT NULL,
    AdressID varchar NOT NULL,
    ShipDate date NOT NULL,
    ShipMode varchar NOT NULL,
    FOREIGN KEY (CustomerID) REFERENCES Customers(CustomerID),
    FOREIGN KEY (AdressID) REFERENCES Adress(AdressID)
);"""
)

c.execute("""
CREATE TABLE Category (
    CategoryID varchar PRIMARY KEY NOT NULL,
    Category varchar(20) NOT NULL,
    Subcategory varchar(20) NOT NULL
);"""
)

```

```python
Cat.to_sql('Category', db_conn, if_exists='append', index=False)
```

```python
df = df.rename(columns={'Sub-Category': 'SubCategory'})
```

```python
df.info()
```

```python
df = pd.merge(df, Cat, how="left", on=["Category", "SubCategory"])
```

```python
df.tail()
```

```python
def divide_columns(row):
    if row['Discount'] != 0:
        return row['Sales'] / (1- row['Discount']) / row["Quantity"]
    else:
        return row['Sales'] / row["Quantity"]
```

```python
df["CataloguePrice"] = df.apply(divide_columns, axis=1)
```

```python
df.tail()
```

```python
df["SupplierPrice"]= (df['Sales'] - df['Profit']) / df["Quantity"]
```

```python
df.head()
```

```python
Pro = pd.DataFrame(columns=["ProductID", "CategoryID", "CataloguePrice", "SupplierPrice", "ProductName"])
```

```python
# Create a product mapping
product_mapping = {product: idx + 1 for idx, product in enumerate(df['Product Name'].unique())}

# Map the 'Product' column to create a 'ProductID' column
df['ProductID'] = df['Product Name'].map(product_mapping)
```

```python
df.head()
```

```python
Pro[["ProductID", "CategoryID", "CataloguePrice", "SupplierPrice", "ProductName"]] = df[["ProductID", "CategoryID", "CataloguePrice", "SupplierPrice", "Product Name"]]
```

```python
Pro = Pro.drop_duplicates()
```

```python
Pro = Pro.drop_duplicates(subset=['ProductID', 'CategoryID'], keep='last')
```
